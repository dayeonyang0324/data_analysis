### matplotlib를 이용한 그래프 그리기

#### plt.plot을 이용하여 값을 설정하는데 리스트를 하나만 넣어주면 y축의 값들로 지정된다
#### 리스트를 2개 넣어주면 첫번째 값은 x축, 두번째 값은 y축이 된다

```python
import matplotlib.pyplot as plt
plt.plot([10, 20, 30, 40])  # 기본값 y축
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [12, 43, 25, 15]) # x축, y축
```

#### plt.title을 통해 표의 제목을 넣을 수 있다

```python
plt.plot([10, 20, 30, 40])
plt.title('plotting')  # 제목 넣기
plt.show()
```

#### 그래프의 선이 2개 이상 존재하는 경우 구분을 위해 범례를 표시한다
- plt.plot에 lable='제목' 값을 넣어주고
- plt.legend()를 이용해 범례를 표에 표시해준다.

```python
plt.title('label')
plt.plot([10, 20, 30, 40], label='asc')  # 범례의 이름을 label로 표현
plt.plot([40, 30, 20, 10], label='desc')
plt.legend()  # legend 호출을 통해 범례를 볼 수 있음
plt.show()
```

#### 범례 색을 설정할 수 있다.

```python
plt.title('color')
plt.plot([10, 20, 30, 40], label='pink', color='pink')
plt.plot([40, 30, 20, 10], 'skyblue', label='skyblue')  # color를 선언하지 않으면 label보다 먼저 와야한다
plt.legend()
plt.show()
```

```python
plt.title('linestyle')
plt.plot([10, 20, 30, 40], label='dashed', color='pink', linestyle='--')
plt.plot([40, 30, 20, 10], 'skyblue', label='dotted', ls=':') 
plt.legend()
plt.show()
```

```python
plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label='circle')
plt.plot([40, 30, 20, 10], 'g^', label='triangle') 
plt.legend()
plt.show()
```
