### csv파일을 시각화하기

- 최고 기온 데이터 그래프

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)  # 데이터 파일의 헤더 삭제

result = []
for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.title('temp')
plt.plot(result, 'r')
plt.show()
```

- 8월의 최고 기온 그래프 그리기

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':  # 매년 8월의 날씨 데이터만 추출
            result.append(float(row[-1]))

plt.title('temp')
plt.plot(result, 'r')
plt.show()
```

- 매년 2월 14일의 최고 기온 그래프

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
            result.append(float(row[-1]))
plt.title('0214 temp')
plt.plot(result)
plt.show()
```

- 매년 2월 14일의 최고, 최저 기온 그래프

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

low = []
high = []
for row in data:
    if row[-1] != '':
        if int(row[0].split('-')[0]) > 1983:
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.title('0214 temp')
# plt.plot(high, low)  # 이렇게 같이 쓰면 안됨
plt.plot(high, 'r')
plt.plot(low, 'b')
plt.show()
```
