#### 주사위 시뮬레이션

```python
import random

dice = []
for i in range(5):
    dice.append(random.randint(1, 6))
    
plt.hist(dice, bins=6)  # bins는 가로축의 갯수
plt.show()
```

```python
import random

dice = []
for i in range(100):  # 많이 반복할 수록 확률은 같으므로 빈도는 비슷해짐
    dice.append(random.randint(1, 6))
    
plt.hist(dice, bins=6)
plt.show()
```

### 기온 데이터를 히스토그램으로 나타내기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []
for row in data:
    if row[-1] != '':
        if 1907 <= int(row[0].split('-')[0]) <= 2017:
            result.append(float(row[-1]))
            
plt.hist(result, bins = 100)
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []
for row in data:
    if row[-1] != '':
        if int(row[0].split('-')[1]) == 8:
            result.append(float(row[-1]))

plt.hist(result, bins = 100)
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []
for row in data:
    if row[-1] != '':
        if int(row[0].split('-')[1]) == 8:
            aug.append(float(row[-1]))
        if int(row[0].split('-')[1]) == 1:
            jan.append(float(row[-1]))

plt.hist(aug, bins = 100, color = 'red')
plt.hist(jan, bins = 100, color = 'blue')
plt.show()
```
