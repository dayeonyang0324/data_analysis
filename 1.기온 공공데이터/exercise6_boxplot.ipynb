### 상자 수염 그림

- 상자 수염 그림이란 최댓값, 3/4Q, 중앙값, 1/4Q, 최솟값을 표현한 값
- 동그라미가 표현된다면 이상값

```python
import random
import matplotlib.pyplot as plt

result = []
for i in range(13):
    result.append(random.randint(1, 1000))

plt.boxplot(result)
plt.show()
```

- 서울 최고 기온 상자 그림

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []
for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
        
plt.boxplot(result)
plt.show()
```

- 8월과 1월의 최고기온 상자그림

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []
for row in data:
    month = int(row[0].split('-')[1])
    if row[-1] != '':
        if month == 8:
            aug.append(float(row[-1]))
        if month == 1:
            jan.append(float(row[-1]))
        
plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```

```python
plt.boxplot(aug, jan)
plt.show()
```

```python
plt.boxplot([aug, jan])
plt.show()
```

#### month별 기온 상자 그림

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

month = [[] for _ in range(12)]
for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1]) - 1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
```

#### 8월의 day별 기온 상자 그림

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = [[] for _ in range(31)]
for row in data:
    if row[-1] != '':
        if int(row[0].split('-')[1]) == 8:
            day[int(row[0].split('-')[2]) - 1].append(float(row[-1]))
        
plt.style.use('ggplot')  # 그래프 스타일 지정
plt.figure(figsize=(10, 5), dpi=300)  # 그래프 크기 수정
plt.boxplot(day, showfliers=False)  # 이상값 생략
plt.show()
```
