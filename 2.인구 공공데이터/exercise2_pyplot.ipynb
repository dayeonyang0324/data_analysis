## 특정 '동'의 인구수 그래프 찾기

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름을 입력해주세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))
    
plt.style.use('ggplot')  # 격자 모양 
plt.rc('font', family='Malgun Gothic')  # 그래프 디자인

plt.plot(result)
plt.title(f'{name} 인구 구조입니다')
plt.show()
```
