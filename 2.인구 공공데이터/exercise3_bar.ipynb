## 막대그래프 그리기
- 히스토그램 : 전체 집단의 분포 상태를 파악하거나 비교할 때 사용된다.
- plt.hist([전체 집단], bins=(가로축 갯수))
- 막대그래프 : 각 항목에 대한 수량의 크기를 정확히 나타낼 수 있다.
- plt.bar([가로축 성분], [해당 가로축의 막대 크기])

```python
import matplotlib.pyplot as plt

# 막대의 위치, 해당 위치의 막대 크기
plt.bar([0, 1, 2, 4, 6, 10], [1, 2, 3, 5, 6, 7])
plt.show()
```

```python
plt.bar(range(6), [2, 3, 4, 5, 7, 8])
plt.show()
```

### 인구 데이터를 막대 그래프로 나타내기

```python
import csv
import matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)

result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))
        
plt.bar(range(101), result)
plt.show()
```

### 막대그래프를 수평으로 나타내기 : barh()

```python
plt.barh(range(101), result)
plt.show()
```

## 성별 인구 데이터

### 연속되는 자료에서 남녀 나눠 그래프 그리기

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

# 남성의 데이터 1~103까지 그 이후에 여성의 데이터가 나옴
male = []
female = []
for row in data:
    if '신도림' in row[0]:
        for i in range(0, 101):
            male.append(int(row[i+3]))
            female.append(int(row[-(i+1)]))
female.reverse()  # 뒤에서부터 append되었으므로 다시 뒤집어준다.
```

```python
import matplotlib.pyplot as plt

plt.barh(range(101), male)
plt.barh(range(101), female)
plt.show()
```

### 항아리 그래프 그리기
- 한 집단을 음수로 변환

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

male = []
female = []
for row in data:
    if '신도림' in row[0]:
        for i in range(0, 101):
            male.append(-int(row[i+3]))  # 그래프가 겹치지 않도록 음수값을 넣어 저장
            female.append(int(row[-(i+1)]))
female.reverse()

import matplotlib.pyplot as plt

plt.barh(range(101), male, label = 'male')
plt.barh(range(101), female, label = 'female')
plt.legend()
plt.show()
```

## 항아리 그래프 주의 사항
- 제주도의 경우 제주 특별 자치도가 포함되는 단어가 46개이므로 정확한 판단이 어렵다.
- 자료를 활용하기 앞서 ','를 모두 삭제해주어야 오류가 나지 않는다.

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

name = input('찾고 싶은 지역의 이름은? : ')

male = []
female = []
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for j in row[106:]:
            female.append(int(j))            
        break


import matplotlib.pyplot as plt

plt.barh(range(101), male, label = 'male')
plt.barh(range(101), female, label = 'female')
plt.legend()
plt.show()
```
