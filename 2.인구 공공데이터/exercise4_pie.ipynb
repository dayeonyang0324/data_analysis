## 파이 차트 그리기

### 파이 차트 함수 : pie()

```python
import matplotlib.pyplot as plt
plt.pie([10, 20])
plt.show()
```

```python
circle = [2442, 4543, 6423, 1260]
plt.axis('equal')  # 파이 차트를 더 동그랗게 하기 위한 명령어
plt.pie(circle)
plt.show()
```

### 레이블 및 범례 추가하기
- labels로 라벨 만들기
- colors로 색 지정해주기
- autopct로 백분율 나타내기
- explode 특정 영역 부각시키기

```python
plt.rc('font', family = 'Malgun Gothic')  # 글씨체 변경
size = [1235, 1643, 2546, 2604]
label = ['A형', 'B형', 'AB형', 'O형']  # 라벨을 달아 표시하기
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
plt.axis('equal')
plt.pie(size, labels = label, colors = color, autopct='%.1f%%', explode=(0, 0, 0.1, 0))  # 소숫점 1자리까지 표시, 차트 부각시키기
plt.legend()
plt.show()
```

## 성별 인구 비율 나타내기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

size = []
name = input('찾고 싶은 지역의 이름은? : ')
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)

# 파이차트 만들기
plt.rc('font', family = 'Malgun Gothic')
color = ['skyblue', 'crimson']
plt.axis('equal')

plt.pie(size, labels = ['남', '여'], autopct='%.1f%%', colors=color, startangle = 90)  # 시작 각도 90일때 12시 방향에서 시작
plt.title(f'{name} 지역의 남녀 성별 비율')
plt.legend()
plt.show()
```
