### 남녀 연령대별 성별 비율

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in  data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break

import matplotlib.pyplot as plt
plt.plot(m, label = 'Male')
plt.plot(f, label = 'Female')
plt.legend()
plt.show()
```

- 연령대별 남성 인구 - 여성 인구

```python
import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
result = []
name = input('궁금한 동네를 입력해주세요 : ')

for row in data:
    if name in row[0]:
        for i in range(3, 104):
            result.append(int(row[i])-int(row[i+103]))
        break
plt.bar(range(101), result)
plt.show()
```

### 산점도 : scatter()
- 가로축과 세로축을 기준으로 두 요소가 서로 어떤 관계를 맺고 있는지 파악하기 쉽게 나타낸 그래프

```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [10, 20, 30, 40])  # 가로축, 세로축
plt.show()
```

#### 버블차트
- s : 점의 크기, c : 점의 색

```python
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3, 4], [10, 20, 30, 40], s=[100, 200, 250, 300], c=['red', 'blue', 'green', 'gold'])
plt.show()
```

#### 컬러바

```python
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3, 4], [10, 20, 30, 40], s=[30, 60, 90, 120], c=range(4))  # c를 범위로 설정해주면 각각 다른색으로 표현됨
plt.colorbar()  # 컬러바
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3, 4], [10, 20, 30, 40], s=[30, 60, 90, 120], c=range(4), cmap='jet')  # cmap 컬러바 색 변경
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=0.7)  # alpha 투명도
plt.colorbar()
plt.show()
```

### 산점도를 이용한 제주도 연령대별 성별 비율

```python
import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
name = input('궁금한 동네를 입력해주세요')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break  # 제주가 포함된 데이터 row가 많으므로 1회만 실시

# 그래프 스타일
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title(f'{name} 지역의 성별 인구 그래프')
# 산점도 그리기
plt.scatter(m, f, c = range(101), alpha=0.5, cmap = 'jet')
plt.colorbar()
# 추세선 추가
plt.plot(range(max(m)), range(max(m)), 'g')  # 추세선 추가
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
