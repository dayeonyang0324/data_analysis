### Html로 불러오기 : read_html

```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table', header=0, index_col=0)
df[1]
```

### 행, 열 슬라이싱

```python
# iloc의 , 앞은 행 뒤는 열에 접근하는 슬라이싱
summer = df[1].iloc[:, :5]  # 모든 행, 앞에서 5개 열 슬라이싱
summer
```

### 열 이름 변경

```python
summer.columns = ['경기수', '금', '은', '동', '계']  # 열의 이름 변경
print(summer)
```

### 특정 값으로 정렬
- 엑셀 파일로 저장하기

```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table', header=0, index_col=0)
summer = df[1].iloc[:, :5]  # 모든 행, 앞에서 5개 열 슬라이싱
summer.columns = ['경기수', '금', '은', '동', '계']  # 열의 이름 변경
summer.sort_values('금', ascending=False).to_excel('하계올림픽메달.xlsx')  # 금을 기준으로 내림차순 정렬
# excel 파일 생성(pip install openpyxl 설치 필요)
```

### periods

```python
import pandas as pd
index = pd.date_range('1/1/2000', periods=8)  # 기간이 8 즉, 1일부터 8일
print(index)
```

### 행, 열 만들기

```python
import numpy as np
df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))  # 0~1사이 8행 3열개의 무작위 데이터 추출, 열은 ABC
df
```

### 필요한 데이터 추출

```python
print(df['B'])  # B열 추출
```

```python
print(df['B'] > 0.4)  # B열에서 0.4보다 큰 값들을 bool형태로 나타내기
```

```python
df2 = df[df['B'] > 0.4]  # True에 해당하는 행들만 갖고 새로운 df만들기
df2
df2.T  # 전치행렬
```

### 행 우선 계산, 열 우선 계산 : axis

```python
import numpy as np
import pandas as pd
index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))
df['D'] = df['A'] / df['B']  # 각 열을 나눠서 새로운 열에 저장
df['E'] = np.sum(df, axis=1)  # 행 우선 계산으로 행의 합 구함 (axis=1 : 열방향)
df = df.sub(df['A'], axis=0)  # 열 우선 계산으로 전체 데이터를 A열의 값으로 뺌 (axis=0 : 행방향)
df = df.div(df['C'], axis=0)  # C열을 기준으로 열 우선 계산 = C열에 해당하는 모든 값
df.head()  # 5개의 행만 출력
```

- csv파일로 저장하기

```python
df.to_csv('text.csv')  # 해당폴더에 csv파일 생성
```
