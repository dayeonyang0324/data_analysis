### 넷플릭스 영화 추천 시스템

```python
import pandas as pd 
import numpy as np

movies=pd.read_csv("movies.csv")  # movie_id, title, genres
ratings=pd.read_csv("ratings.csv")  # user_id, movie_id, ratings
```

```python
# 필요 데이터들만 병합
data=pd.merge(ratings, movies, on="movieId")
column=['userId','movieId','rating','title','genres']
data=data[column]
data
 
movie_rating=data.pivot_table(index="title",  columns='userId')['rating']
movie_rating.head()
```

```python
# NaN은 -1값으로 변경
movie_rating.fillna(0, inplace=True)
movie_rating
```

```python
movie_rating.shape
```

### 유사도 기반 추천 시스템

**추천 시스템은 content-base와 collaborative filtering으로 나눌 수 있다.**
- content-base : 추천하고자 하는 아이템에 대한 정보를 기반으로 추천하는 알고리즘. (예: 특정 배우, 장르)
- collaborative filtering : 사용자 간 유사도에 중점을 둔 추천 시스템, 사용자의 행동 패턴에 따라 추천이 가능함.

**collaborative filtering**
- item-based CF : 상품 간 유사도를 기반으로 추천
- user-based CF : 유저 간 유사도를 기반으로 상품을 추천
- matrix factorization CF : 행렬 기반 추천 시스템

#### 피어슨 상관계수 : matrix factorization / SVD(특이값 분해)
- 두 변수 X 와 Y 간의 선형 상관 관계를 계량화한 수치로서  -1 ~ 1 사이의 값을 가진다.
- +1은 완벽한 양의 선형 상관 관계, 0은 선형 상관 관계 없음, -1은 완벽한 음의 선형 상관 관계를 의미한다.
- 일반적으로 상관관계는 피어슨 상관관계를 의미하는 상관계수이다.
- 상관계수 = 공분산 / (표준편차 * 표준편차)

```python
from sklearn.decomposition import TruncatedSVD

syd = TruncatedSVD(n_components = 12)  # latent 값 12
matrix = syd.fit_transform(movie_rating)
matrix.shape
```

```python
# 피어슨 상관계수 구하기
corr = np.corrcoef(matrix)
corr.shape
```

```python
# heatmap 그리기
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.heatmap(movie_rating)
```

```python
movie_title = movie_rating.columns
title_list = list(movie_title)
coffey = title_list.index('Toy Story (1995)')
```

```python
corr_coffey = corr[coffey]
list(movie_title[(corr_coffey >= 0.9)])[:10]
```

#### 코사인 유사도 : item based CF
- 각도 기반의 계산법으로 벡터의 크기에 영향을 받지 않는다.
- 값의 범위는 -1 ~ 1이며 1에 가까울수록 유사하다고 판단한다.

```python
# item_based (pip install sklearn)
from sklearn.metrics.pairwise import cosine_similarity

item_based = cosine_similarity(movie_rating)
item_based = pd.DataFrame(data = item_based, index = movie_rating.index, columns = movie_rating.index)
item_based.head()
```

```python
# 1에 가까운 유사도 찾기
def recommend(title):
    # 제목과 유사한 영화 상위 5개 추천
    return item_based[title].sort_values(ascending=False)[:6]
```

```python
recommend('Toy Story (1995)')
```
